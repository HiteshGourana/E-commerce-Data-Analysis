```python
df.groupby(['zone','category'])['revenue'].sum()
```

```python
l=(df.groupby(['category','zone'])['revenue'].sum().reset_index())
```

```python
l
```

```python
l.shape
```

```python
plt.pie(l.iloc[0:5,2],labels=l.iloc[0:5,1],autopct="%.2f%%")
plt.title(l.iloc[:1,0])
plt.show()
```

```python
k=0
j=5
plt.figure(figsize=(18, 8))
for i in range(8):
    plt.subplot(2,4,i+1)
    values = l.iloc[k:j, 2].values  
    max_val = max(values)
    explode = [0.6 if v == max_val else 0 for v in values]
    plt.pie(l.iloc[k:j,2],labels=l.iloc[k:j,1],autopct="%.2f%%",explode=explode,shadow=True)
    plt.title(l.iloc[k,0],fontweight='bold')
    k=k+5
    j=j+5
plt.tight_layout()
plt.suptitle("Category Wise Sales", fontsize=16, fontweight="bold")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
l2=(df.groupby('category')['customer_gender'].value_counts().reset_index())
```

```python
l2
```

```python
l2.sort_values(['category', 'customer_gender'])
```

```python
pivot = l2.pivot(index='category',
                 columns='customer_gender',
                 values='count')
pivot
```

```python
plt.figure(figsize=(18, 8))

for i in range(len(pivot)):
    plt.subplot(2, 4, i+1)
    
    plt.pie(
        pivot.iloc[i].values,
        labels=pivot.columns,
        autopct="%.2f%%",
        shadow=True,
        startangle=90
    )
    
    plt.title(pivot.index[i], fontweight='bold')

plt.suptitle("Category Wise Sales (Male vs Female)",
             fontsize=16, fontweight="bold")

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```

```python
l3=(df.groupby(['state','category'])['units_sold'].sum().reset_index())
```

```python
l3=l3.sort_values('units_sold',ascending=False)
```

```python
l3
```

```python
name=list(l3.iloc[0:,1].unique())
name
l3[l3['category']==name[0]]
```

```python
plt.figure(figsize=(18, 8))
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.bar(l3[l3['category']==name[i]].iloc[0:,0],l3[l3['category']==name[i]].iloc[0:,2])
    plt.title(l3[l3['category']==name[i]].iloc[i,1],fontweight="bold")
    plt.xticks(rotation=90)

plt.suptitle("Country Wise Sales", fontsize=16, fontweight="bold")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
